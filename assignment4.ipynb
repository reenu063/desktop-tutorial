```python
#select Country
unique_Country = raw_data["Country"].drop_duplicates()
```

```python
#Print list of Countries
print("unique_Country")
```

```python
#select specific Country (United Kingdom)
Country_df = raw_data[raw_data["Country"] == "Canada"]
```

```python
print(Country_df)
```

```python
#selecting the associated column with mental health
Country_df_hist = Country_df[["Age", "Gender", "treatment", "mental_health_consequence"]]
```

```python
#Print Gender 
print(Country_df_hist.Gender.unique())
```

```python
#Print Age 
print(Country_df_hist.Age.unique())
```
